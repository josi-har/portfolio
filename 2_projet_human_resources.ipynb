```python
### Comme les variables quantitatives présentent des échelles hétérogènes (ex. : salaire vs heures travaillées),
#ce qui peut fausser la lecture des visualisations dans cette boxplot. Pour harmoniser leur représentation,
# une standardisation a été appliquée via StandardScaler de sklearn pour qu'on voit bien
# 1. les Tendance centraleles (moyenne, médiane, mode)
# 2.dispersion et 3. Outliers, 4.Symétrie ou asymétrie
from sklearn.preprocessing import StandardScaler

###  Créer le scaler, standardscaler = standardise les variables numériques en leur donnant(une moyenne = 0,un écart-type = 1)
#Évite qu’une variable avec des grandes valeurs domine le plus petite
scaler = StandardScaler()
### chaque colonne est transformé avec fit_transform(fit=calcul, transform = transformer), pd.DataFrame = converti le résultat en dataframe
df_scaled = pd.DataFrame(scaler.fit_transform(df[colonnes_quantitatives.columns]), columns=colonnes_quantitatives.columns)

plt.figure(figsize=(14, 4))
### les donnes = dataframe transformé

sns.boxplot(data=df_scaled)
plt.title("Boxplot des variables quantitatives standardisées", fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
### Afficher la distribution des variables quantitatives boxplot(boît à moustache)

### len donne le nombre des colonnes quantitatives, n = variable qui contienne le nombre des colonne qunatitiatives
n = len(colonnes_quantitatives.columns)

###personnalisé les couleurs
colors = ['#8e44ad',  '#d35400', '#3498db', '#2ecc71', '#e74c3c', '#f1c40f', '#1abc9c']

### fig =figure global, axes = un tableau numphy 2D de 2 x 4 (2 ligne et 4 éléments)
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
###flatten =transform le tableau 2D en 1D(plus pratique)
axes = axes.flatten()

### Commancé le boucle , i = num de colonne, col = nom de colonne, enumerate = parcour la liste de colonne quantitative
for i, col in enumerate(colonnes_quantitatives.columns):

    ###  axe x vont être les colonnes quantitatives, couleur c'est numéro des colonnes,
    #axes[i] = chaque colonne quantitative est représentée par son propre boxplot, placé dans un des sous-graphes de la grille (axes)
    sns.boxplot(x=colonnes_quantitatives[col], color=colors[i], ax=axes[i])

    ###utilisation de f-string pour donner les titres au graphiques
    axes[i].set_title(f"Distribution de {col}")

# Supprimer les axes vides (le 8e dans ce cas)
### range (n, len(axes)) = génère les nombres de n jusqu’à len(axes)-1 : n = le nombre de colonne quanti, len(axes) = nombre de sous-plot, -1=le dernier élément de la liste
for j in range(n, len(axes)):


    ### fig.delaxes = supprime le sous-plot vide
    fig.delaxes(axes[j])


### ajuste les espacements entre les sous-graphes pour éviter que Les titres ou labels se chevaucher
plt.tight_layout()
plt.show()
```

```python
### Afficher la distribution des variables quantitatives avec histplot

### n variable qui contienne le nombre des colonne quanti, .columns =donne la liste des noms des colonnes
n = len(colonnes_quantitatives.columns)


### créer un tableau de 3 lignes 3 éléments
fig, axes = plt.subplots(3, 3, figsize=(20, 12))
colors = ['#8e44ad',  '#d35400', '#3498db', '#2ecc71', '#e74c3c', '#f1c40f', '#1abc9c']

### transformer tableau 2D en 1D
axes = axes.flatten()


### Utiliser les axes pour tracer chaque graphique
for i, col in enumerate(colonnes_quantitatives.columns):

    ### x = les colonnes quanti, couleur = chosi dan sla liste,  subplot où le graphique doit apparaître
    ### kde=True (ajoute une courbe de densité (Kernel Density Estimation))
    sns.histplot(x=colonnes_quantitatives[col], kde=True, color = colors[i], ax=axes[i])


    axes[i].set_title(f"Distribution de {col}")

# Supprimer les axes vides (le 8e dans ce cas)
for j in range(n, len(axes)):
    fig.delaxes(axes[j])


### ajuste les espacements entre les sous-graphes pour éviter que Les titres ou labels se chevaucher
plt.tight_layout()
plt.show()
```

# Règle des 1.5 × IQR pour détecter les outliers
### IQR = Q3- Q1 (intervalle interquartile)

### outliers = toutes les valeurs:
##Inférieures à:     𝑄1-1.5 x IQR,
##Supérieures à:     𝑄3+1.5 x IQR        
###Ces valeurs sont trop éloignées du cœur de la distribution

```python
### calculer manuellement  des outliers dans les salaires avec des  quartille Q1, Q3 et règle des 1.5 * IQR
Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
IQR = Q3 - Q1
outliers_salary = df[(df["Salary"] < Q1 - 1.5 * IQR) | (df["Salary"] > Q3 + 1.5 * IQR)]
outliers_salary.sort_values(by="Salary")
```

```python
### calculer manuellement  des outliers dans les enquetes d'engagement avec des  quartille Q1, Q3 et règle des 1.5 * IQR
Q1 = df["EngagementSurvey"].quantile(0.25)
Q3 = df["EngagementSurvey"].quantile(0.75)
IQR = Q3 - Q1
outliers_engagementsurvey = df[(df["EngagementSurvey"] < Q1 - 1.5 * IQR) | (df["EngagementSurvey"] > Q3 + 1.5 * IQR)]
outliers_engagementsurvey .sort_values(by="EngagementSurvey")
```

# 6.8. Exploration de variable Salaire et sa corrélation avec la  performance


```python
###le salaire et le variable le plus dispersé parmit les autres varibales
sns.violinplot(x=df["Salary"])
```

```python
### : C’est une forte asymétrie (3.30), ce qui signifie que : La majorité des salaires sont concentrés vers les bas niveaux.
#Il existe quelques salaires très élevés qui tirent la moyenne vers le haut.
# La moyenne est supérieure à la médiane
###La plupart des employés gagnent des salaires relativement modestes, mais une minorité touche des salaires très élevés.
#Cela crée une distribution déséquilibrée, avec une longue traîne vers les hauts salaires.
### Transformation logarithmique : Pour réduire l’asymétrie et faciliter l’analyse.
#Utiliser la médiane plutôt que la moyenne pour représenter le salaire typique

###coefficient d’asymétrie (skewness)
df["Salary"].skew()
```

```python
### Kurtosis = 15.45 → très élevée. Cela signifie que la distribution des salaires contient distribution leptokurtique
### Forte concentration autour de la moyenne : la majorité des salaires sont proches de la moyenne.
# Une distribution leptokurtique présente les caractéristiques suivantes
# Présence de valeurs extrêmes (outliers) : quelques salaires très éloignés (très élevés ici) tirent la distribution vers les extrêmes.
#Queue épaisse : les extrémités de la distribution sont plus lourdes que dans une distribution normale


### kurtosis (ou coefficient d’aplatissement) de la distribution des salaires
df["Salary"].kurt()
```

```python
#transformer la variable Salary en échelle logarithmique, ce qui est très utile en analyse statistique
#Après transformation logarithmique, l’asymétrie est nettement réduite. de 3.30 à 1.63
#La distribution devient plus équilibrée, ce qui améliore la fiabilité des corrélations, des régressions et des visualisations
Salary_log = np.log(df["Salary"] + 1)
df["Salary_log"] = Salary_log
```
